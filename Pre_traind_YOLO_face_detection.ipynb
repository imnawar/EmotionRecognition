````python
from yoloface.face_detector import YoloDetector
import numpy as np
from PIL import Image

model = YoloDetector(target_size=720, device="cpu", min_face=90)
orgimg = np.array(Image.open('test_image.jpg'))
bboxes,points = model.predict(orgimg)
````

````python
bboxes,points
````

````python
import cv2
````

````python
bboxes[0][0]
````

````python
from google.colab.patches import cv2_imshow
````

````python
x1, y1, x2, y2 = int(bboxes[0][0][0]), int(bboxes[0][0][1]), int(bboxes[0][0][2]), int(bboxes[0][0][3]) ## get x, y of the bbox

cv2.rectangle(orgimg, (x1, y1), (x2, y2), (0, 255, 0), 2) ## create bbox rectangle
cv2.putText(orgimg, 'Face', (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 3) ## create bbox text

cv2_imshow(orgimg) ## show frame
````
