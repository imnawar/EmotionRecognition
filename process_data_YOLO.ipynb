```python
from zipfile import ZipFile
from pathlib import Path
import os
```

```python
def extract_data(zipfile_path: Path, destination_path: Path) -> None:
    '''Extracts zipfile'''

    if destination_path.is_dir():
        print(f"{destination_path} exists.")
    else:
        print(f"{destination_path} doesn't exist, creating one...")
        destination_path.mkdir(parents=True, exist_ok=True)

    if not os.listdir(destination_path):
        with ZipFile(zipfile_path, 'r') as zip:
            print("Extracting files...")
            zip.extractall(destination_path)

            print("Extracting finished.")
    else:
        print("Data already extracted.")

data_path = Path("/content")
zipfile_path = data_path / "/content/drive/MyDrive/proccesed_data.zip"
images_path = data_path / "proccesed_data"

extract_data(zipfile_path, images_path)
```

```python
for index, row in df.iterrows():
    class_label = row['encoded_label']
    boxx = row['boxx'][0]
    row_content = " ".join([str(class_label), str(boxx[0]), str(boxx[1]), str(boxx[2]), str(boxx[3])])
    file_name = row['img_num'].split('.')[0]

    file_path = "/".join(row['new_path'].split('/')[:-1]) +'/'+ file_name + ".txt"
    f = open(file_path, "w")
    f.write(row_content)
    f.close()
    print(file_path)
```
